```python
import re
from itertools import product
from collections import ChainMap
```

```python
def mask_mem(num,mask):
    temp = "".join([j if j=='X' else i if j=='0' else '1' for i,j in zip(format(num, '036b'),mask)])
    Xs = temp.count('X')
    comb = list(product(['0','1'], repeat=Xs) )
    
    mems = [int(''.join([i[0]+i[1] for i in zip(temp.split('X'),j+tuple(['']))]),2) for j in comb]
    
    return mems
```

```python
mem = {}
for i in data:
    l_dic = [{j:k[1] for j in mask_mem(k[0],i[0])} for k in i[1:]]
    l_dic.reverse()
    temp = dict(ChainMap(* l_dic ))
    
    mem={**mem, **temp}
```
