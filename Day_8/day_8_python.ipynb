# Part 1

```python
def navigate(instructions, acc = 0, ind = [0]):
    
    if len(ind) != len(set(ind)):
        return acc
    
    elif instructions[ind[-1]][0]=='acc':
        return navigate(instructions, acc + instructions[ind[-1]][1], ind+[ind[-1]+1])
    
    elif instructions[ind[-1]][0]=='jmp':
        return navigate(instructions, acc, ind+[ind[-1]+instructions[ind[-1]][1]])
        
    else:
        return navigate(instructions, acc , ind+[ind[-1]+1])
```

# Part 2

```python
def navigate(instructions, acc = 0, ind = [0]):
    
    if len(instructions) == ind[-1]+1:
        if instructions[ind[-1]][0]=='acc':
            acc += instructions[ind[-1]][1]
        return [acc,"END"]
    
    elif len(ind) != len(set(ind)):
        return [acc,"LOOP"]
    
    elif instructions[ind[-1]][0]=='acc':
        return navigate(instructions, acc + instructions[ind[-1]][1], ind+[ind[-1]+1])
    
    elif instructions[ind[-1]][0]=='jmp':
        return navigate(instructions, acc, ind+[ind[-1]+instructions[ind[-1]][1]])
        
    else:
        return navigate(instructions, acc , ind+[ind[-1]+1])
```
